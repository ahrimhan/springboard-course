## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

### 1. Top 10 countries with most projects

We use `json_normalize` to flatten semi-structure data further. This is essential because each country can have multiple projects.

```python
# load json as string
js1 = json.load((open('data/world_bank_projects.json')))
#df1 = json_normalize(js1, 'mjtheme_namecode', ['code', 'name'], errors='ignore', record_prefix='_')
df1 = json_normalize(js1, 'mjtheme_namecode','countryname')
df1.head()
```

```python
df1.shape
```

```python
df1.groupby('countryname').size().sort_values(ascending=False).head(10)
```

### 2. Top 10 major project themes

We can obtain top 1p major project themes by using `groupby` 'code' or 'name'. However, there are some entries that have only the code and name is missing.

```python
df1.groupby('name').size().sort_values(ascending=False).head(10)
```

```python
df1.groupby('code').size().sort_values(ascending=False).head(10)
```

### 3. Filling missing values of 'name'

1) We make the dataframe (`code_name_map`) indicating the map matching 'code' and 'name' of projects.

```python
import numpy as np

#df2 = df1[['code', 'name']]
df2 = df1.copy()

df2.drop("countryname", axis=1, inplace=True)
#To use dropna(), we replace empty entry to np.nan
df2['name'].replace('', np.nan, inplace=True)
df2.dropna(subset=['name'], inplace=True)

#converted 'code' to numeric type
#df2.code = df2.code.astype(int)
df2['code'] = pd.to_numeric(df2['code'])

code_name_map = df2.drop_duplicates().sort_values('code').reset_index(drop=True)
```

```python
code_name_map
```

Or, we can construct the dict (`code_name_map2`) for mapping 'code' and 'name' of projects.

```python
code_name_map2 = {}

for index in np.arange(len(df1)):
    c = pd.to_numeric(df1.code[index])
    if c not in code_name_map2.keys():
        code_name_map2[c] = df1.name[index]
```

```python
code_name_map2
```

2) To handle with the **missing values**, we fill the empty names referencing the map (`code_name_map`) obtained above.

```python
for index, row in df1.iterrows():
    
    #if name is missing value
    if df1.name[index] == '':
        df1.name[index] = code_name_map.name[pd.to_numeric(df1.code[index]) - 1]
```

```python
df1.head(10)
```

3) Therefore, we filled the missing entries of names, so the **final result of the top 10 major project themes** are as follows.

```python
df1.groupby('name').size().sort_values(ascending=False).head(10)
```
